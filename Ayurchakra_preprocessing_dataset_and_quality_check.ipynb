```python
import pandas as pd
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the CSV file
df = pd.read_csv("prakriti_dataset.csv")

# Extracting features (x) and labels (y)
x = df.iloc[:, :-2].values.tolist()
y = df.iloc[:, -2].tolist()

# Lowercasing
x = [' '.join(map(str.lower, sublist)) for sublist in x]

# Removal of special characters and symbols other than alphabets and numbers
def clean_text(text):
    text = re.sub(r'[^a-z0-9\s]', '', text)
    text = ' '.join(text.split())
    return text

x = [clean_text(text) for text in x]

# Tokenization
tokenized_x = [word_tokenize(text) for text in x]

# Stopwords Removal
stop_words = set(stopwords.words('english'))
no_stopwords_x = [[word for word in sublist if word not in stop_words] for sublist in tokenized_x]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_x = [[lemmatizer.lemmatize(word) for word in sub_list] for sub_list in no_stopwords_x]

# Vectorization using TF-IDF
tfidf = TfidfVectorizer()
x_tfidf = tfidf.fit_transform([' '.join(sublist) for sublist in lemmatized_x])

# Label Encoding
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x_tfidf, y_encoded, test_size=0.3, shuffle=True, random_state=42)

# K-Nearest Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)

# Predictions for KNN on the training set
predictions_train_knn = knn.predict(x_train)
print("KNN  Accuracy:", accuracy_score(predictions_train_knn, y_train))



# Support Vector Machine (SVM) Classifier
svm = SVC(kernel='linear')
svm.fit(x_train, y_train)

# Predictions for SVM on the training set
predictions_train_svm = svm.predict(x_train)
print("SVM Accuracy:", accuracy_score(predictions_train_svm, y_train))
```
